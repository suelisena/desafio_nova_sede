```python
#Plotar dados de Recife
dados = {
    'Cidade de Destino': ['Natal (RN)', 'Fortaleza (CE)', 'Maceió (AL)', 'João Pessoa (PB)', 'Teresina (PI)'],
    'Tempo de Entrega (h)': [4, 10, 3, 2, 14]
}

df = pd.DataFrame(dados)

# Gerar o gráfico de barras
plt.figure(figsize=(10, 6))
plt.bar(df['Cidade de Destino'], df['Tempo de Entrega (h)'], color='skyblue')

# Adicionar títulos e rótulos
plt.title('Tempo de Entrega de Recife para Capitais do Nordeste', fontsize=16)
plt.xlabel('Cidade de Destino', fontsize=12)
plt.ylabel('Tempo de Entrega (h)', fontsize=12)

# Adicionar o tempo de entrega em cada barra para melhor visualização
for i, tempo in enumerate(df['Tempo de Entrega (h)']):
    plt.text(i, tempo + 0.5, str(tempo), ha='center', va='bottom')

plt.xticks(rotation=45, ha='right')  # Rotaciona os nomes das cidades para evitar sobreposição
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Dados dos tempos de entrega a partir de Recife para as capitais do Nordeste
dados_recife = {
    'Cidade de Destino': ['Natal (RN)', 'Fortaleza (CE)', 'Maceió (AL)', 'João Pessoa (PB)', 'Teresina (PI)', 'São Luís (MA)'],
    'Tempo de Entrega (h)': [4, 10, 3, 2, 14, 18]
}
df_recife = pd.DataFrame(dados_recife)

# Dados dos tempos de entrega a partir de Salvador para as capitais do Nordeste
dados_salvador = {
    'Cidade de Destino': ['Natal (RN)', 'Fortaleza (CE)', 'Maceió (AL)', 'João Pessoa (PB)', 'Teresina (PI)', 'São Luís (MA)'],
    'Tempo de Entrega (h)': [15, 22, 12, 13, 20, 24]
}
df_salvador = pd.DataFrame(dados_salvador)

# Calcular a média dos tempos de entrega para cada cidade
media_recife = df_recife['Tempo de Entrega (h)'].mean()
media_salvador = df_salvador['Tempo de Entrega (h)'].mean()

# Criar um DataFrame com as médias para o gráfico
df_medias = pd.DataFrame({
    'Cidade': ['Recife', 'Salvador'],
    'Média de Tempo de Entrega (h)': [media_recife, media_salvador]
})

# Imprimir os valores das médias
print("Média de tempo de entrega:")
print(f"Recife: {media_recife:.2f} horas")
print(f"Salvador: {media_salvador:.2f} horas")

# Gerar o gráfico de barras comparativo
plt.figure(figsize=(8, 6))
df_medias.set_index('Cidade').plot(kind='bar', color=['#007ACC', '#FFC300'])

plt.title('Comparativo da Média de Tempo de Entrega para Capitais do Nordeste')
plt.ylabel('Média de Tempo de Entrega (h)')
plt.xlabel('Cidades')
plt.xticks(rotation=0)
plt.legend(title='Localização', loc='best')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
